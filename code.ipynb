```python
import matplotlib.pyplot as plt
import numpy as np
from os.path import exists

```

```python
import seaborn as sns
```

```python
def add_file():
    data = [None] * 8

    for i in range(0, 8):
        ascii_output_file =  "detector" + str(i+1) + ".txt"
        if(exists(ascii_output_file)):
            data[i] = np.loadtxt(ascii_output_file)
    return data

data = add_file()

```

```python
def fig_plot(fig, axes, data):
    fig.suptitle("Gaussian beam with sigmaX = sigmaY = -5", fontsize="16")
    fig.supxlabel("x position")
    fig.supylabel("y position")
    plot(axes[0], data[0])
    axes[0].set_title("Detector 1 (4mm away from beam source)")

    plot(axes[1], data[7])
    axes[1].set_title("Detector 8 (5915mm away from beam source)")


def plot(axes,data,type = None):
    x = data[:,0]
    y= data[:,1]
    axes.scatter(x,y)

fig1, axes = plt.subplots(2, sharex=True, sharey=True)
fig_plot(fig1, axes, data)
plt.savefig('pics/gaussian_report.png', dpi=1200)
plt.show()
```
